```python
import pandas as pd
import mooodule

# 연습문제
# - 육군신체 측정정보 데이터를 이용해서 BMI지수를 확인해보자
# - 데이터 오픈 > 컬럼정보를 확인 > 데이터 전처리 > BMI 계산
# BMI지수 = 몸무게(kg) / 키(m) ** 2
```

```python
# 1. 데이터 불러오기
data = pd.read_csv('./data/육군신체측정정보.csv', encoding='utf-8')
data.head()
```

```python
# 2. 데이터 확인
data.info()

# 각 컬럼에 어떤것들이 들어있는지, 결측치를 확인할 수 있다.
# 확인해야 하는것 -> 
#   1) 결측치가 존재하는가?
#   2) 각 컬럼의 Dtype이 어떻게 구성되어 있는가?
# 문제점: BMI를 구하기 위해서는 숫자데이터의 연산이 필요하다
# 키, 몸무게의 type이 object로 되어있어서 형변환이 필요하다 -> float형
""" 
RangeIndex 개의 인덱스를 가진다
순번은 결측치가 없다
가슴둘레는 1개의 결측치
머리둘레는 2개의 결측치를 가지고있다.
"""
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
# 3. 필요한 컬럼 추출

# height
h = data['신장 센티미터']
# weight
w = data['몸무게 킬로그램']
```

```python
# 4. 단위 제거 -> 수치형으로 데이터타입 변경
h = h.str.replace('cm', '').str.strip().astype(float)
w = w.str.replace('kg', '').str.strip().astype(float)

# .str 으로 문자에 접근
```

```python
# 5. 키(cm)를 키(m)로, 단위 변환하기
h = h/100
h
```

```python
# 6. BMI 지수 계산하기

bmi = w / h ** 2
```

```python
# BMI 컬럼의 BMI데이터 추가하기
data['BMI'] = bmi
```

```python
# 큰값에서 작은값을 빼주는 내가 만든 함수 max_min 정의
def max_min(x):
    return x.max() - x.min()
```

```python
### 카테고리 생성하기

# 나이 데이터로 카테고리 생성하기
ages = [0,2,10,15,21,23,37,61,20,41,100,32]
```

```python
# 범주 -> 맨 앞 데이터 지정을 주의해야함.
# 왼쪽의 값은 포함되지 않고, 오른쪽의 값은 포함된다.

# 0세 ~ 19세 : 미성년자
# 20세 ~ 35세 : 청년
# 36세 ~ 60세 : 중년
# 61세 ~ : 노년

# 0은 포함 안되지만, 20세는 포함됨.

bins = [0, 19, 35, 60, 100]
labels = ['미성년자', '청년', '중년', '노년']

# pd.cut() 함수를 이용해서 범주형 데이터로 변환하기
# 연속적인 데이터를 구간으로 나누는데 사용되는 함수
age_categories = pd.cut(ages, bins=bins, labels=labels, right=True)
age_categories
```

```python
# age 데이터들의 카테고리화 된 각각의 결과들
# 데이터의 길이 (개수)
# 카테고리의 개수
```

```python
age_df = pd.DataFrame(ages, columns=['age'])
age_df
```

```python
age_df['age_category'] = pd.cut(age_df['age'], bins=bins, labels=labels, right=True)
age_df
```
