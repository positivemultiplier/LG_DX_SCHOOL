### 네이버 영화리뷰 분석

```python
# 텍스트 상태의 데이터가 아닌 파이썬 객체 자체를 저장하고 불러오는 방법
import pickle
```

```python
# pickle 모듈을 이용하여 데이터 로드
with open('movie_review.pkl', 'rb') as f :
    all_review = pickle.load(f)
```

```python
all_review
```

- 전체 리뷰 개수 확인

```python
print(len(all_review))
```

- 리뷰 길이 평균 값

```python
re_total_len = 0

for review in all_review :
    re_total_len += len(review[0])

re_avg_len = re_total_len/len(all_review)
print(f"리뷰의 평균 길이는 '{re_avg_len}'입니다.")
```

- 긍정리뷰/부정리뷰 개수

```python
pos_re = 0
neg_re = 0

for review in all_review :
    if review[1] == 1:
        pos_re +=1
    elif review[1] == 0 :
        neg_re +=1
        
print(f"긍정 '{pos_re}'개")
print(f"부정 '{neg_re}’개")
```

```python
review_pn = []
for review in all_review :
    review_pn.append(review[1])
    
pos_re = review_pn.count(1)
neg_re = review_pn.count(0)

print(f"긍정 '{pos_re}'개")
print(f"부정 '{neg_re}’개")
```

- 리뷰길이 최대 값, 리뷰 내용

```python
max_content = all_review[0][0]
for review in all_review :
    if len(max_content) < len(review[0]):
        max_content = review[0]
        
print(f"리뷰 최대길이는 '{len(max_content)}'입니다.")
print(max_content)    
```

- 'ㅋ' 문자가 포함된 리뷰 개수

```python
ㅋ_count = 0
for review in all_review :
    if 'ㅋ' in review[0]:
        ㅋ_count +=1
print(f"'ㅋ'가 포함된 리뷰는 {ㅋ_count}개 입니다. ")
```

- '검색데이터' 가 포함된 리뷰 개수

```python
search_data = input('검색할 데이터 입력 >> ')
search_data_count = 0
for review in all_review :
    if search_data in review[0]:
        search_data_count +=1
        print(review[0])
print(f"'{search_data}'가 포함된 리뷰는 {search_data_count}개 입니다. ")
```
