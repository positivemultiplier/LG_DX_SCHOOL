```python
from selenium import webdriver as wd 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 파일시스템을 위한 라이브러리(파일, 폴더를 생성하고 삭제하는 라이브러리)
import os # 바탕화면에 폴더를 만들때 사용

# 이미지의 링크를 파일로 변환하는 라이브러리 (retrieve : 회수하다. 다시찾다, 만회하다  )
from urllib.request import urlretrieve
```

```python
# 01 폴더 생성
# 직관적으로 대화를 해야한다.
# 대화 : 만약에 바탕화면에 이미지라는 폴더가 없다면, 바탕화면에 이미지 폴더를 만들어라.
# 부정의 조건문 if not os.path.exists("C:/Users/lgdx/Desktop/이미지"):

# public 폴더는 관리자권한으로 접근이 불가능하다. lgdx는 관리자권한이 있는 계정이다.
# 역슬러시는 운영체제가 인식한느것이다. 사용자는 슬래시로 사용한다.


# 바탕화면에 이미지 폴더가 없다면, 바탕화면에 이미지 폴더를 생성한다.
if not os.path.isdir("C:/Users/lgdx/Desktop/이미지") :
    # 바탕화면에 이미지 폴더를 생성한다.
    os.mkdir("C:/Users/lgdx/Desktop/이미지")
    # 폴더를 삭제한다.
    # os.rmdir("C:/Users/lgdx/Desktop/이미지") 
```

```python
driver = wd.Chrome()
driver.get("https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=lg")

```

```python
# 이미지를 불러오기위해 무한 스크롤 실행 !!! 로딩을 다 한 상태에서 selenium은 진행해야한다. 
body = driver.find_element(By.CSS_SELECTOR,"body")
while True :
    first = driver.page_source
    body.send_keys(Keys.END)
    time.sleep(1)
    second = driver.page_source
    if(first == second) :
        break
```

```python
# 이미지 태그 수집 후, src 값만 저장
image = driver.find_elements(By.CSS_SELECTOR, "#main_pack > section > div.api_subject_bx._fe_image_tab_grid_root.ani_fadein > div > div > div.image_tile._fe_image_tab_grid > div > div > div > div > img")
image
```

```python
url_list = []
for i in image:
    url_list.append(i.get_attribute("src"))
url_list
```

```python
# 수집한 url을 실제 파일로 변환
for i in range(len(url_list)):
    # 첫번째 매개변수는 url_list[i]의 값을 가져온다.
    # 두번째 매개변수는 저장할 파일의 경로와 이름을 지정=> format을 사용하여 파일이름을 지정할 수 있다.(f"C:/Users/lgdx/Desktop/이미지/image{i}.jpg") 
    # 
    urlretrieve(url_list[i],f"C:/Users/lgdx/Desktop/이미지/image{i}.jpg")
```
