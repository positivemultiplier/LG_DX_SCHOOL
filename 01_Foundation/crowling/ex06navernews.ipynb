```python
# selenium 패키지의 webdriver 모듈을 wd라는 이름으로 import
# By, Keys 모듈을 각각 By, Keys라는 이름으로 import
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
#1. 크롬 브라우저 실행
driver = wd.Chrome()
```

```python
#2. 네이버사이트로 이동
driver.get("https://www.naver.com")
```

```python
# 3. 검색창에 "태풍" 검색 후, 엔터치기
search = driver.find_element(By.CSS_SELECTOR,"#query")
search.send_keys("태풍")
search.send_keys(Keys.ENTER)
```

```python
# 4. 네비게이션바에서 뉴스 클릭하기 => anchor 태그를 찾아서 클릭해야한다. 
btn = driver.find_element(By.CSS_SELECTOR,"#lnb > div.lnb_group > div > div.lnb_nav_area._nav_area_root > div > div.api_flicking_wrap._conveyer_root > div:nth-child(8) > a")
btn.click()
```

```python
# 6. 뉴스제목만 추출 후, 컨텐츠만 출력해보기 
#fdr-96afdd69af484b2284889f3f9d215da6 > div > div > div > div:nth-child(1) > div.sds-comps-base-layout.sds-comps-full-layout.j0ZmtUURh3xcbwC9nntE > div > a.Q4Y2t_v6TP0bwxdR9avD.CL8ortS8q0ByUMOIzqrc > span
title = driver.find_elements(By.CSS_SELECTOR,"#fdr-83e079b4a93a498ab1aee80d0fce01e5 > div > div > div > div > div.sds-comps-base-layout.sds-comps-full-layout.j0ZmtUURh3xcbwC9nntE > div > a.Q4Y2t_v6TP0bwxdR9avD.CL8ortS8q0ByUMOIzqrc > span")
for i in title :
    print(i.text)
```

```python
# 6.스크롤 내리기 => 비동기라 body를 다시 찾을 필요가 없다. 
body = driver.find_element(By.CSS_SELECTOR,"body")

for i in range(100):
    body.send_keys(Keys.END)

```

```python
title = driver.find_elements(By.CSS_SELECTOR, "div > div > div > div > div.sds-comps-base-layout.sds-comps-full-layout.j0ZmtUURh3xcbwC9nntE > div > a.Q4Y2t_v6TP0bwxdR9avD.CL8ortS8q0ByUMOIzqrc > span ")
for i in title:
    print(i.text)
```

## 선택자를 복사할때 최근 기법
- 복사한 선택자에 의미를 알 수 없는 난수의 id난 class가 존재
- 대규모의 포털에서 사용하는 정책
- 난수가 포함된 아이디(#)는 지워버리자.
- 클래스(.)는 지우지 말고, 의미있는 클래스만 남기자.  

```python
# 링크 수집 Selenium
# a태그를 수집한 후 , href 송성을 추출
#fdr-910bb1a0898f4800b04a3034e726898d > div > div > div > div:nth-child(9) > div.sds-comps-base-layout.sds-comps-full-layout.X0Vnt5k9cEAZ0Md6SrX8 > a
aTag = driver.find_elements(By.CSS_SELECTOR,"div > div > div > div > div > div > a")
aTag
# 한솥도시락(사람처럼 코딩하는 방법, 오류가 났을 때 대처하는 방법)
```

```python
# Selenium에서 속성을 수잡할 때는 get_attribute("속성명") 
# aTag 안에 있는 속성을 가지고오기. 
aTag[0].get_attribute("href")
```
