#### 문자열 다루는 함수(기능)

````python
# count('문자') : 문자열에 포함된 '문자'의 개수를 세주는 기능
s = "Life is too short, You need 'python'"
# 문자열 안에 'e'문자는 몇번 들었을까?
s.count('e') # ->3
````

````python
# 'Life' 몇번 들었을까?
s.count('Life') # -> 1
s.count('life') # -> 0 소문자, 대문자 구별하여 인식
````

````python
# 파이썬 문자열 함수 중에서 소문자로 변경하는 기능
#lower() : 파이썬 문자열 함수 중에서 소문자로 변경하는 기능
s_lower = s.lower()
print(s_lower)

# upper(): 소문자를 대문자로 바꾸는 기능(함수)
s_upper =s_lower.upper()
print(s_upper)
print(s_lower)
print(s)
s_upper.count('S')
````

````python
# 소문자로 입력해서 life 몇번 들었는지 찾기!
print(s.lower().count('Life'))
print(s_lower.count('life'))
print(s_upper.lower().count('life'))
````

````python
s.lower().count('l')
````

````python
# strip() : 양쪽 공백을 제거하는 함수
s_space = "   안녕!   "
print(s_space)
s_space_strip = s_space.strip()
print(s_space_strip)

# lstrip() : 왼쪽 공백 제거
# rstrip() : 오른쪽 공백 제거

print('왼쪽:', s_space.lstrip())
print('오른쪽:', s_space.rstrip())
print('양쪽:', s_space.strip())
````

````python
# 문자열 찾아서 위치 알려주는 기능 : find('문자'), index('문자')
s_space_strip[-1]
s_space_strip[2]
str_idx = s_space_strip.find('!') # -> 2 인덱스 번호 반환
s_space_strip[str_idx]

````

````python
# 예제) 문자열에서 날짜와 날씨를 출력하기
day = "2025년 7월 1일의 날씨는 흐림입니다."
start_idx = day.find('2025년')
# print(start_idx)
end_idx = day.find('일')
# print(end_idx)
d = day[start_idx : end_idx+1] #find 함수로 찾기!!

w = day[-6:-4]
print('날짜:', d )
print('날씨:', w )

````

````python
# 과제) 문자열 관련 함수 중에 특정 문자를 찾아서 다른걸로 대체하는 기능
# 문자열 찾아서 바꾸는 기능의 이름은? 
day.replace('2025', '200000000000')
````
