```python
def get_grade(x):
    grade = ""
    avg = x.mean()
    if avg >= 90:
        grade = "A"
    elif avg >= 80:
        grade= "B"
    elif avg >= 70:
        grade = "C"
    elif avg >= 60:
        grade = "D"
    else:
        grade = "F"
    return grade

data["grade"] = data.iloc[:,2:5].apply(get_grade, axis=1)
data
```
