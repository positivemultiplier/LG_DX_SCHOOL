## pandas `.drop()` 메서드 매개변수

| 매개변수 | 설명 | 타입 | 기본값 |
|---------|------|------|-------|
| **labels** | 삭제할 인덱스 또는 컬럼명 | single label or list-like | None |
| **axis** | 삭제할 축 (0=행, 1=열) | {0 or 'index', 1 or 'columns'} | 0 |
| **index** | 삭제할 행 인덱스 | single label or list-like | None |
| **columns** | 삭제할 컬럼명 | single label or list-like | None |
| **inplace** | 원본 수정 여부 | bool | False |
| **errors** | 오류 처리 방식 | {'ignore', 'raise'} | 'raise' |

### 사용 예시

```python
# 행 삭제
df.drop('행인덱스', axis=0)
df.drop(index='행인덱스')

# 열 삭제  
df.drop('컬럼명', axis=1)
df.drop(columns='컬럼명')

# 여러 개 삭제
df.drop(['행1', '행2'], axis=0)
df.drop(['열1', '열2'], axis=1)

# 원본 수정
df.drop('컬럼명', axis=1, inplace=True)

# 오류 무시
df.drop('존재하지않는컬럼', axis=1, errors='ignore')
```

**핵심 포인트:**
- `axis=0` (기본값): 행 삭제
- `axis=1`: 열 삭제  
- `inplace=False` (기본값): 원본 유지, 새로운 DataFrame 반환
- `inplace=True`: 원본 DataFrame 직접 수정

````python
# 포켓몬 실습
data = {'피카츄':9631,'꼬부기':3393,

        '파이리':1490,'이상해씨':2632}
 


poketmon = pd.Series(data)
poketmon
 
````

````python
# 레벨업한 데이터가 들어있는 poketmon_up Series
data2 = {'피카츄':9904,'꼬부기':3448,

         '이브이':2466,'뮤':2890}

poketmon_up = pd.Series(data2)
print(poketmon_up)

print("-"*100)
````

````python
# 포켓몬의 올라간 능력치를 계산해보자!
# 인덱스명이 달라졌을따 Series는 어떻게 연산을 할까? => key값이 같지 않다면 NaN(Not a Number 결측치(없는값)) cf 이상치()

attack = poketmon_up - poketmon
attack
````

````python
# 비어있는 데이터 보고싶지 않다면? .notnull() = > output 은 boolean 형식으로 나온다.
# 보여줄수 있는것은 True 아닌것은 False
a = attack.notnull()
print(a)
````

````python
# boolean indexing 하면된다. 
b = attack[attack.notnull()]
print(b)
````

````python
# 비어있는 데이터만 보고싶다면 ? .isnull()
c = attack[attack.isnull()]
print(c)
````

````python
# 레벨업해서 올라간 공격력의 증가율 계산해보기! (NaN 표시되지 않게) .notnull()
# 
up_rate = (poketmon_up - poketmon) / poketmon
print(up_rate[up_rate.notnull()])
````
