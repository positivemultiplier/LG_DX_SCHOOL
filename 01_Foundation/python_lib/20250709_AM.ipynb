### 넘파이 배열의 인덱싱과 슬라이싱
#### indexting
- 인덱싱, 슬라이싱
#### slicing
#### boolean indexing
- 불리언 인덱싱 -> 논리연산자를 활용하여 True인 값을 출력할 수 있도록 데이터에 접근
- 조건에 맞는 데이터에 접근

```python
# 데이터 접근을 위한 indexing  vs slicing
# Quize 출력해보기 : 10,20,30,40,50

# 1st 기본적인 방법 arr4 = np.array([10,20,30,40,50])
arr4 = np.array([10,20,30,40,50])

# 2nd numpy의 arange function(start, end, step) 활용하는 방법
arr4 = np.arange(10, 51, 10)

arr4

```

```python
# 원하는 값을 가지고와서 출력하기 ! 
# indexing => 배열명[index number]
# 1st : 첫번재 숫자[0] 가지고와서(indexing) 출력하기(printing) 
# step1 : indexing
a = arr4[0]
# step2 : printing
print(a)

# 2nd : 첫번째와 세번째 숫자를 가지고와서(indexing) 출력하기(printing) 
# 10, 30 데이터 인덱싱 하고싶다 => index number 0,2
# 여러개의 데이터에 접근하고싶다면 대괄호로 한번 더 감싸준다
# step1 : indexing
b = arr4[[0,2]]
# step2 : printing
print(b)

# 3rd : 첫번째부터 5번째 숫자를 가지고와서(slicing) 출력하기(printing)
# 배열명[row ][start index num(default : 0) : end index num(default : unlimit ) : step (default : +1)]
# step1 : slicing
c = arr4[0:5:1]
# step2 : printing
print(arr4[0:5:1])

# 4th : 조건에 맞는 것들만 가지고와서(boolean indexing) 출력하기(printing)
# 불리언인덱싱 => 배열명[T/F 배열] => True 값만 출력
# True 값만 가져온다. 
# step1 : boolean indexting
d = arr4[arr4 > 30]
# step2 : printing
print(d)


# asdjkhasd
```

```python
# 문제 11~20까지 배열을 만들어서 제곱의 값이 200 이상인 숫자만 출력

arr5 = np.arange(11,21,1)
base = arr5
print(base)

a = arr5 ** 2 >= 200
print(a)

# boolean indexing [ ] 
b = arr5[arr5 ** 2 >= 200]
print(b)

```
