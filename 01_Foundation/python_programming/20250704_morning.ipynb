```python
first_num = 1
second_num = 1

while True:
    first_num = int(input("    >>"))
    print(f"    >>{first_num}")
    second_num = int(input("    >>"))
    print(f"    >>{second_num}")
    
    
    sum = first_num + second_num
    
    if first_num == 0 and second_num == 0:
        print(" 0  .")
        break
    else:
        print("   : {}".format(sum))    
```

```python
#      
#     0  .

while True:
    #    
    first_num = int(input("    >> "))
    
    #    
    second_num = int(input("    >> "))
    
    #     0
    if first_num == 0 and second_num == 0:
        print("   0 .  .")
        break  # while 
    
    #    
    sum = first_num + second_num
    
    #  
    print("  : {}".format(sum))
```
