```python
com_num = random.randint(1, 50)

choice_num = 0

while com_num != choice_num:  
    choice_num = int(input("1~50 사이의 숫자를 입력하세요:"))
    print(f"1~50 사이의 숫자를 입력하세요: {choice_num} ")    
    if com_num < choice_num:
        print(f"{choice_num} 보다 작은 수 입니다")
    elif com_num > choice_num:
        print(f"{choice_num} 보다 큰 수 입니다.")
   
print(f"{choice_num} 정답을 맞추셨습니다.")   
```

```python
import random
com_num = random.randint(1, 50)  # 1~50 사이의 랜덤 숫자 생성
print('(정답 확인용) 컴퓨터 숫자:', com_num)  # 실제 게임에서는 주석처리 가능
while True:
    user_input = input('숫자를 입력하세요 >> ')
    if not user_input.strip().isdigit():
        print('숫자를 올바르게 입력하세요.')
        continue
    input_num = int(user_input)
    if com_num > input_num:
        print(f'{input_num} 보다 큰 수 입니다.')
    elif com_num < input_num:
        print(f'{input_num} 보다 작은 수 입니다.')
    else:
        print('정답을 맞추셨습니다!')
        break
```

```python
list_ice = ["탱크보이", "수박바","엄마는외계인"]
cnt = 1
for i in list_ice:
    print(cnt, i)
    cnt += 1
```

```python
score_list = [90, 45, 70, 60, 55]
idx = 1

for i in score_list:
    if i >= 60:
        print( f"{idx}합격입니다.")
    else :
        print(f"{idx}불합격입니다.")
    idx += 1                                                                
    
```

```python
# 정보처리기사 자격증 점수 리스트
scores = [90, 45, 70, 60, 55]  # 60점 이상이면 합격
student_number = 1  # 학생 번호를 1번부터 시작
for score in scores:
    # 점수가 60점 이상이면 합격, 아니면 불합격 출력
    if score >= 60:
        print(str(student_number) + "번 학생은 합격입니다.")
    else:
        print(str(student_number) + "번 학생은 불합격입니다.")
    student_number = student_number + 1  # 다음 학생 번호로 증가
```

```python
score_list = [90, 45, 70, 60, 55]
cnt = 1
check = ''
for i in score_list:
    if i >= 60:
        check = "합격"
    else:
        check = "불합격"
    print(f"{cnt}번 학생은 {check}입니다")
    cnt += 1
```

```python
# e 리스트 반복시키면서 리스트 요소 출력하기
# 2. 사람의 이름만 출력하기
# 3. name_list = [] 리스트 선언한 후에 해당 리스트에 각 사람의 이름 누적 저장하기!
# 리스트 선언
e = [[1, '손지원'], [2, '임예희'], [3, '정민지']]
name_list = []

for i in e:
    # i[1] 값이 리스트에 추가하는 기능 실행
    # i[1] 값이 리스트에 추가하는 기능 실행 : 대상.append(값)
    name_list.append(i[1])
print(name_list)
    


```

```python
# 숫자를 입력 받고 입력 받은 숫자의 약수를 구하시오.
#(약수란 어떤 수를 나누어 떨어지게 하는 수)
divisors = []
num = int(input("약수를 구할 숫자를 입력하세요: "))
for i in range(1, num + 1):
    if num % i == 0:
        divisors.append(i)
print(f"{num}의 약수는: {divisors} 입니다.")
```
