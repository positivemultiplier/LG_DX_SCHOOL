#### 숫자 1개를 입력 받을때 약수를 출력하는 함수를 정의하시오.
- 

```python
# 변수생성
input_num = int(input("약수를 구할 숫자 입력 >> "))

# 함수 정의 
def divisor(num):
    """약수를 구하는 함수"""
    ida = []
    for i in range(1, num + 1):
        if num % i == 0:
           ida.append(i)
    return ida 

# 함수 호출
divisor(input_num) 



             
```

#### 함수의 return 키워드 이해하기

```python
# 더하고, 빼는 결과를 반환하는 함수 정의
def add_sub(num1, num2):
    return num1 + num2 , num1 - num2
```

```python
# 사용자 정의 함수 호출하기 (사용하기)
add_sub(4, 3) # add_sub(num1 = 4, num2 = 3)

# (): 튜플
# 함수 return 반환값을 언제나 한개만 반환함
# packing 패킹 됨
```

```python
add, sub = add_sub(4, 3) # (7, 1) add 에 7, sub 에 1이 들어감
print(add) # unpacking  언패킹 됨
print(sub)
```

#### 함수의 매개변수 기본값 설정

```python
# 목표 : 지수연산을 하는 함수 생성하기 
# 지수연산자 : **2 
def power_num(num, power=2):
    """지수연산을 하는 함수"""
    return num ** power
    
```

```python
# 2의 2승 구하기
# power=2 기본값 설정이 되어 있기 때문에
power_num(num = 3)
# num = 3 입력하지 않아도, 3만 입력하더라도 num의 값으로 대입된다. 기본값 default value
```

```python

# 전역변수 선언
num1 = 1
num2 = 99

# 함수 정의
def cal_num(num1, num2):
    return num1 + num2, num1 - num2, num1 * num2, num1/num2


```

```python
# 사용자로부터 두 숫자 입력 받기
num11 = int(input("첫 번째 숫자를 입력하세요: "))
num12 = int(input("두 번째 숫자를 입력하세요: "))

# 함수 호출
result_add, result_sub, result_mul, result_div = cal_num(num11, num12)
print(f"덧셈: {result_add}")
print(f"뺄셈: {result_sub}")
print(f"곱셈: {result_mul}")
print(f"나눗셈: {result_div}")
```
