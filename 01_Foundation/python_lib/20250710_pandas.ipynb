## üìà 2. Series (ÏãúÎ¶¨Ï¶à) ÏôÑÏ†Ñ Ïù¥Ìï¥

### SeriesÏùò ÌäπÏßï
- **1Ï∞®Ïõê Î∞∞Ïó¥** ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞ Íµ¨Ï°∞
- **Ïù∏Îç±Ïä§**ÏôÄ **Í∞í**Ïù¥ **1:1 Îß§Ïπ≠**
- DictionaryÏ≤òÎüº **indexÎ•º key**Î°ú ÏÇ¨Ïö©


````python
# NumPy vs Pandas Series ÎπÑÍµê
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("üîç NumPy Î∞∞Ïó¥ vs Pandas Series ÎπÑÍµê")
print("="*50)

# NumPy Î∞∞Ïó¥
np_array = np.array([90, 85, 88])
print(f"NumPy Î∞∞Ïó¥: {np_array}")
print(f"ÌÉÄÏûÖ: {type(np_array)}")

print("\n" + "="*50)

# Pandas Series
pd_series = pd.Series([90, 85, 88], index=["Íµ≠Ïñ¥", "ÏòÅÏñ¥", "ÏàòÌïô"])
print(f"Pandas Series:")
print(pd_series)
print(f"ÌÉÄÏûÖ: {type(pd_series)}")
print(f"Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ: {pd_series.dtype}")

# ÏãúÍ∞ÅÌôî ÎπÑÍµê
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# NumPy Î∞∞Ïó¥ ÏãúÍ∞ÅÌôî
ax1.bar(range(len(np_array)), np_array, color='lightcoral', alpha=0.7)
ax1.set_title('üìä NumPy Î∞∞Ïó¥ (Ïù∏Îç±Ïä§ Î≤àÌò∏)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Ïù∏Îç±Ïä§')
ax1.set_ylabel('Ï†êÏàò')
ax1.set_xticks(range(len(np_array)))
ax1.set_xticklabels(['0', '1', '2'])

# Pandas Series ÏãúÍ∞ÅÌôî
ax2.bar(pd_series.index, pd_series.values, color='skyblue', alpha=0.7)
ax2.set_title('üìä Pandas Series (ÎùºÎ≤® Ïù∏Îç±Ïä§)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Í≥ºÎ™©')
ax2.set_ylabel('Ï†êÏàò')

plt.tight_layout()
plt.show()

print("\nüí° ÌïµÏã¨ Ï∞®Ïù¥Ï†ê:")
print("   NumPy: Ïù∏Îç±Ïä§ Î≤àÌò∏Î°úÎßå Ï†ëÍ∑º Í∞ÄÎä•")
print("   Pandas: ÏùòÎØ∏ÏûàÎäî ÎùºÎ≤®Î°ú Ï†ëÍ∑º Í∞ÄÎä•")
````

## üõ†Ô∏è 3. Series ÏÉùÏÑ± Î∞©Î≤ïÎì§

### Îã§ÏñëÌïú ÏÉùÏÑ± Î∞©Î≤ï
1. **Î¶¨Ïä§Ìä∏**Î•º ÏÇ¨Ïö©Ìïú ÏÉùÏÑ±
2. **ÎîïÏÖîÎÑàÎ¶¨**Î•º ÏÇ¨Ïö©Ìïú ÏÉùÏÑ±
3. **Ïù∏Îç±Ïä§ ÏßÄÏ†ï**ÌïòÏó¨ ÏÉùÏÑ±


````python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Í∏∞Î≥∏ Series ÏÉùÏÑ± (ÏûêÎèô Ïù∏Îç±Ïä§)
team_members = pd.Series([
    "ÏÑúÌù¨Ï∞Ω", "Î∞ïÏÑ±Ìôî", "Ïã¨ÏàòÏßÑ", 
    "Ïù¥ÏßÑÌóå", "Ïù¥Í¥ëÎØº", "Ïù¥ÏùÄÏßÄ"
])
print("=== Í∏∞Î≥∏ Series (ÏûêÎèô Ïù∏Îç±Ïä§) ===")
print(team_members)
print(f"ÌÉÄÏûÖ: {type(team_members)}")

# 2. Ïù∏Îç±Ïä§Î•º ÏßÄÏ†ïÌïú Series ÏÉùÏÑ±
team_mbti = pd.Series([
    "ISTJ", "ENTJ", "ISFJ", 
    "INTJ", "ENTP", "ISTP"
], index=[
    "ÏÑúÌù¨Ï∞Ω", "Î∞ïÏÑ±Ìôî", "Ïã¨ÏàòÏßÑ", 
    "Ïù¥ÏßÑÌóå", "Ïù¥Í¥ëÎØº", "Ïù¥ÏùÄÏßÄ"
])
print("\n=== Ïù∏Îç±Ïä§ ÏßÄÏ†ï Series ===")
print(team_mbti)

# 3. Î≥ÄÏàòÎ•º Ïù∏Îç±Ïä§Î°ú ÏÇ¨Ïö©
team_mbti_v2 = pd.Series([
    "ISTJ", "ENTJ", "ISFJ", 
    "INTJ", "ENTP", "ISTP"
], index=team_members)
print("\n=== Î≥ÄÏàòÎ•º Ïù∏Îç±Ïä§Î°ú ÏÇ¨Ïö© ===")
print(team_mbti_v2)

# üéØ Ïã§Ïäµ: ÌåÄÏõê Ï†ïÎ≥¥Î°ú Series ÎßåÎì§Í∏∞
print("üë• ÌåÄÏõê Ï†ïÎ≥¥ Series ÏÉùÏÑ±")
print("="*40)

# ÌåÄÏõê Ïù¥Î¶Ñ Series
team_names = pd.Series([
    "ÏÑúÌù¨Ï∞Ω", "Î∞ïÏÑ±Ìôî", "Ïã¨ÏàòÏßÑ", 
    "Ïù¥ÏßÑÌóå", "Ïù¥Í¥ëÎØº", "Ïù¥ÏùÄÏßÄ"
])

print("üìã ÌåÄÏõê Î™ÖÎã®:")
print(team_names)
print(f"\nüìä Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ: {team_names.dtype}")
print(f"üìè Îç∞Ïù¥ÌÑ∞ Í∞úÏàò: {team_names.size}")

# ÌåÄÏõê Îã¥Îãπ ÏóÖÎ¨¥ Series
team_roles = pd.Series([
    "ÌåÄÏû•", "Ïó∞ÎùΩÎã¥Îãπ", "Î∂ÑÏüÅÏ°∞Ï†ïÎã¥Îãπ",
    "Ïã†ÏÜçÎåÄÏùëÎã¥Îãπ", "Î∞úÌëúÎã¥Îãπ", "ÏÑúÍ∏∞"
], index=team_names)

print(f"\nüè∑Ô∏è ÌåÄÏõêÎ≥Ñ Îã¥Îãπ ÏóÖÎ¨¥:")
print(team_roles)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))

# ÌåÄÏõêÎ≥Ñ Îã¥Îãπ ÏóÖÎ¨¥ ÏãúÍ∞ÅÌôî
plt.subplot(2, 2, 1)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#c2c2f0']
plt.pie([1]*len(team_names), labels=team_names, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('üë• ÌåÄÏõê Íµ¨ÏÑ±', fontsize=14, fontweight='bold')

# Îã¥Îãπ ÏóÖÎ¨¥ Í∏∏Ïù¥ Î∂ÑÏÑù
plt.subplot(2, 2, 2)
role_lengths = team_roles.str.len()
plt.bar(range(len(role_lengths)), role_lengths, color='lightgreen', alpha=0.7)
plt.title('üìè Îã¥Îãπ ÏóÖÎ¨¥ Ïù¥Î¶Ñ Í∏∏Ïù¥', fontsize=14, fontweight='bold')
plt.xlabel('ÌåÄÏõê Î≤àÌò∏')
plt.ylabel('Í∏ÄÏûê Ïàò')

# Îã¥Îãπ ÏóÖÎ¨¥ Î∂ÑÌè¨
plt.subplot(2, 2, 3)
role_counts = team_roles.value_counts()
plt.bar(range(len(role_counts)), role_counts, color='orange', alpha=0.7)
plt.title('üìä Îã¥Îãπ ÏóÖÎ¨¥ Î∂ÑÌè¨', fontsize=14, fontweight='bold')
plt.xlabel('ÏóÖÎ¨¥ Ïú†Ìòï')
plt.ylabel('Ïù∏Ïõê Ïàò')
plt.xticks(range(len(role_counts)), role_counts.index, rotation=45, ha='right')

# ÌåÄÏõê Î≤àÌò∏Î≥Ñ Î∂ÑÌè¨
plt.subplot(2, 2, 4)
member_indices = range(len(team_names))
plt.scatter(member_indices, [1]*len(team_names), s=100, c=colors, alpha=0.7)
for i, name in enumerate(team_names):
    plt.annotate(name, (i, 1), xytext=(5, 5), textcoords='offset points')
plt.title('üìç ÌåÄÏõê Î∂ÑÌè¨', fontsize=14, fontweight='bold')
plt.xlabel('ÌåÄÏõê Î≤àÌò∏')
plt.ylabel('Ï°¥Ïû¨ Ïó¨Î∂Ä')
plt.ylim(0.5, 1.5)

plt.tight_layout()
plt.show()
````

## ‚öôÔ∏è 5. Series Îç∞Ïù¥ÌÑ∞ Ï°∞Ïûë (CRUD)

### CRUD Ïó∞ÏÇ∞
- **Create**: Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä
- **Read**: Îç∞Ïù¥ÌÑ∞ Ï°∞Ìöå (Ïù∏Îç±Ïã±/Ïä¨ÎùºÏù¥Ïã±)
- **Update**: Îç∞Ïù¥ÌÑ∞ ÏàòÏ†ï
- **Delete**: Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú

```mermaid
graph LR
    A[CRUD Ïó∞ÏÇ∞] --> B[Create: Ï∂îÍ∞Ä]
    A --> C[Read: Ï°∞Ìöå]
    A --> D[Update: ÏàòÏ†ï]
    A --> E[Delete: ÏÇ≠Ï†ú]
```

### ÏÇ≠Ï†ú Î∞©Î≤ï ÎπÑÍµê
1. **`del`**: ÏßÅÏ†ë ÏÇ≠Ï†ú (ÏõêÎ≥∏ Î≥ÄÍ≤Ω)
2. **`.drop()`**: ÏûÑÏãú ÏÇ≠Ï†ú (ÎØ∏Î¶¨Î≥¥Í∏∞)
3. **`.drop(inplace=True)`**: ÏòÅÍµ¨ ÏÇ≠Ï†ú

````python
# Ïù∏Íµ¨ Îç∞Ïù¥ÌÑ∞Î°ú CRUD Ïã§Ïäµ
population = pd.Series(
    [10000000, 3000000, 1200000, 2000000],
    index=["ÏÑúÏö∏", "Î∂ÄÏÇ∞", "Í¥ëÏ£º", "ÎåÄÍµ¨"]
)
population.name = "2024ÎÖÑ Ïù∏Íµ¨Ïàò"
population.index.name = "ÎèÑÏãú"

print("=== ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ===")
print(population)

# 1. UPDATE: Îç∞Ïù¥ÌÑ∞ ÏàòÏ†ï
print("\n=== Îç∞Ïù¥ÌÑ∞ ÏàòÏ†ï (Î∂ÄÏÇ∞ Ïù∏Íµ¨) ===")
population["Î∂ÄÏÇ∞"] = 3500000
print(f"ÏàòÏ†ïÎêú Î∂ÄÏÇ∞ Ïù∏Íµ¨: {population['Î∂ÄÏÇ∞']:,}Î™Ö")

# 2. CREATE: Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä
print("\n=== Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä (ÎåÄÏ†Ñ) ===")
population["ÎåÄÏ†Ñ"] = 1500000
print(population)
````

````python
# 3. DELETE: Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú (Ïó¨Îü¨ Î∞©Î≤ï)
print("=== ÏÇ≠Ï†ú Î∞©Î≤ï 1: .drop() ÎØ∏Î¶¨Î≥¥Í∏∞ ===")
preview_drop = population.drop("ÎåÄÏ†Ñ")
print("ÎØ∏Î¶¨Î≥¥Í∏∞:")
print(preview_drop)
print("\nÏõêÎ≥∏ÏùÄ Í∑∏ÎåÄÎ°ú:")
print(population)

print("\n=== ÏÇ≠Ï†ú Î∞©Î≤ï 2: inplace=TrueÎ°ú ÏòÅÍµ¨ ÏÇ≠Ï†ú ===")
population_copy = population.copy()  # Î≥µÏÇ¨Î≥∏ÏúºÎ°ú ÌÖåÏä§Ìä∏
population_copy.drop("ÎåÄÏ†Ñ", inplace=True)
print("ÏòÅÍµ¨ ÏÇ≠Ï†ú Í≤∞Í≥º:")
print(population_copy)

# Ï£ºÏùò: del ÌÇ§ÏõåÎìúÎäî ÏßÅÏ†ë ÏÇ≠Ï†úÌïòÎØÄÎ°ú Ïã†Ï§ëÌïòÍ≤å ÏÇ¨Ïö©
# del population["ÏÑúÏö∏"]  # Ïù¥Î†áÍ≤å ÌïòÎ©¥ Î∞îÎ°ú ÏÇ≠Ï†úÎê®
````
