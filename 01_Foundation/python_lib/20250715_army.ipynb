```python
# phase3.데이터 확인
# 앞으로는 대용량의 데이털르 처리하기때문에 항상 datatype을 확인해야한다.  
# RangeIndex는 index번호다. 16만개가 가지고있다? 
# 가슴 둘레 센치미터는 1개의 결측치 null값이 있다.
# 머리 둘레 센티미터는 2개의 결측치 null값이 있다.
# 데이터 전처리가 필요하다.

data.info()
# 확인해야하는것 => 1. 결측치가 존재하는가? 2. 각 컬럼의 Dtype이 어떻게 구성되어 있는가? 
# 문제점 : BMI를 구하기 위해서는 숫자 데이터의 연산이 필요하다!
# 키, 몸무게 데이터가 object로 되어있어서 형변환이 필요하겠다.! Object-> float 
```

```python
# 필요한 컬럼을 추출 => 열인덱싱
# 키 data 추출
height  = data["신장 센티미터"]
# 몸무게 data 추출
weight = data["몸무게 킬로그램"]
```

```python
# phase4 Data 정제하기 
# 단위제거 => 띄어쓰기부터 _cm를 제거해야한다.

# step1. height변수에 접근 !! 
# 변수의 string타입을 잡아서(sting안잡아도 접근이 되네...) slicing을 통해서 접근한다.
# height.str[ : ]
# weight.str[ : ]

# step2 공백과 _cm를 제거하고 공백으로 대체한다.=> replace()
# height = height.str.replace(" cm", "")


# weight = weight.str.replace(" kg", "")

# step3 수치형으로 데이터타입 변경 =>.astype()
# height = height.astype("float32")
# weight = weight.astype("float32")

```

```python
# step2 공백과 _cm를 제거하고 공백으로 대체한다.=> replace()
height = height.str.replace(" cm", "")
weight = weight.str.replace(" kg", "")
```

```python
# step3 수치형으로 데이터타입 변경 =>.astype()
height = height.astype("float32")
weight = weight.astype("float32")
```

```python
# step4 cm => m로 변환하기 위해서 100으로 나누기
height = height / 100
```

```python
# phase5. BMI 계산하기
# BMI = 몸무게(kg) / (키(m) * 키(m))
BMI = weight / (height ** 2)
BMI
```

```python
# phase6. 데이터프레임으로 만들기
data["BMI"] = BMI
```
