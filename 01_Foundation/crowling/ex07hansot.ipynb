````python
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
import pandas as pd
import time
````

````python
#1. 사이트 요청하기
driver = wd.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
# 서버와 클라이언트가 통신하면 time.sleep()을 필요해다  => 환경에 맞게 설정해야한다. => 통신환경을 고려해야한다.
# 서버가 클라이언트에게 정보를 줄때 필요하다
# 굳이 클라이언트가 서버줄때는 필요 없다. 
# 이미지일경우 sleep(초를 늘려줘야한다. ) 
time.sleep(1)
````

````python
#2. 상품이름, 가격 수집하기
# 포인트 : 특정 요소를 검사할 때는 반드시 개발자도구에서 영역을 확인하자. 
# 문제점 : 요소를 선택하면 a 태그만 선택이 된다.
# 해결책 : 주변 태그를 직접 분석한다. 
# 만들어둔 이유 : 사용자가 영역 어디든지 클릭했을 때 상품페이지를 보여주기 위해서 => 사용자의 편의성을 증진시키기 위함

title = []
price = []


print("상품명 수집하기 ")
# 선택자의 테그를 찾아와서 복수의 elements를 수집한다. 뒤에서부터 단수를 지우자. #menuList_627 > div > div.item-text > h4  // id# 와 nth-child를 지운다. 
title_elements = driver.find_elements(By.CSS_SELECTOR, " div > div.item-text > h4")
for i in title_elements :
     title.append(i.text)
     print(i.text)
print("---" * 10)


print("가격 수집하기")
# 선택자의 테그를 찾아와서 복수의 elements를 수집한다. 뒤에서부터 단수를 지우자. #menuList_627 > div > div.item-text > div > strong  // id# 와 nth-child를 지운다. 
price_elements = driver.find_elements(By.CSS_SELECTOR," div > div.item-text > div > strong")
for i in price_elements :
     price.append(i.text)
     print(i.text) 
print("---" * 10)



# 개수를 확인해야한다
print("반드시 data의 개수를 확인해야한다")
print(len(title_elements),len(price_elements))
````

````python
print(title)
print(price)

data = pd.DataFrame({"상품명": title, "가격": price})
data
````

# 예외처리 
- 코드가 동작하다가 오류가 발생하면 모든 프로그램 중단
- 영역이 2가지 존재 try(오류 나기전까지 시도할 코드) / except(오류가 났을 때 실행할 코드)
- tip! 데이터를 수집할 때 예외처리를 사용한다면 반드시 없을때는 없다라는 의미의 데이터로 처리할 것. 


## 1. Python의 결측값 표현

### ① None
- **Python 표준 내장 자료형에서 결측(비어있음, 정의되지 않음) 상태**를 나타낼 때 사용하는 특수 객체입니다.
- 대소문자 주의: **None** (소문자×)
- 예시:
  ```python
  a = None
  ```

### ② null
- **Python에는 null이라는 예약어/객체가 존재하지 않습니다.**
- null을 입력하면 NameError 발생
  ```python
  a = null  # NameError: name 'null' is not defined
  ```

### ③ NaN (Not a Number)
- 주로 **수치 연산**에서 결측값, 정의되지 않은 숫자 등을 나타냅니다.
- float 타입 및 numpy, pandas 등에서 사용
- 예시
  ```python
  import math, numpy as np
  a = float('nan')  # 표준 파이썬 NaN
  b = np.nan        # numpy의 NaN
  ```
- NaN은 수치연산에서 사용되고, 비교 연산 등에 주의가 필요 (NaN != NaN)

### ④ Na
- **Na**는 파이썬 기본 예약어가 아니며, 변수명이나 NA객체가 따로 정의된 게 아닙니다.
- pandas(판다스) 등 일부 라이브러리에서 결측값을 **pd.NA**로 표기합니다 (버전 1.0이상).
  - 예시:  
    ```python
    import pandas as pd
    s = pd.Series([1, pd.NA, 3])
    ```

## 2. 실무 적용 및 주의사항

- **일반 파이썬 코드나 자료구조에서 결측값(Null/Na/결손 등)을 나타내고 싶으면 반드시 None을 사용**
- **수치 데이터(특히 numpy, pandas)에서 결측/정의불능 데이터를 표현할 때는 NaN(pd.NA 포함)을 사용**
- **null, Na 등은 파이썬 기본 문법에서 지원하지 않으므로 쓸 수 없습니다**

## 3. 회계, 세무, 데이터분석 실무에서 결측값 처리 TIP

- **데이터 입력 및 전처리 단계**에서 결측값이 일관적으로 처리되어야 하며,
  - 리스트, 딕셔너리 등에서는 None
  - 수치형 벡터/Array에서는 np.nan 또는 pd.NA
- **DB/타 언어 연동 시**: SQL/JavaScript 등은 NULL, R은 NA, Python은 None/NaN이 표준임을 기억하고 적절히 변환 필요

## 4. 요약

| 용어      | Python 지원여부 | 용도                             |
|-----------|-----------------|---------------------------------|
| None      | O               | 일반 결측/비어있음                |
| null      | X               | 사용 불가 (NameError)             |
| Na        | X               | 사용 불가 (pandas의 NA는 pd.NA)    |
| NaN       | O (float/numpy) | 수치 결측, 정의불능, pandas 등에서 |

**즉, 파이썬에서는 None(일반) 또는 NaN/NA(수치, 자료분석)만 사용 가능**하며,  
null이나 Na 키워드는 직접적으로 쓸 수 없습니다.

# time 사용법
- 목적 : 코드 중간에 쉬는 시간을 부여할 때 사용한다. 
- 사용법 : 반드시 ★★★요청★★★이 발생하면 무조건 시간을 부여한다.
- 응답시간을 벌어주는것이 핵심이다.
- tip : 정해진 시간은 없다. 환경에 맞춰서 수정하며 사용한다. 보통 1초면 된다.  

````python
# while문으로 더보기 클릭해보기 => 반드시 break문을 생각한 후 진행하자. 
while True : 
    try :
        # 더 많은 데이터를 가져오기 위해서 더보기 클릭.
        btn = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
        # 클릭 실행하기.
        btn.click()
        # 반드시 clock() 후에 wait를 해야한다.    
        time.sleep(1)
    except :
        print("더이상 더보기 버튼이 없습니다. 수집을 종료합니다.")  
        break
````
