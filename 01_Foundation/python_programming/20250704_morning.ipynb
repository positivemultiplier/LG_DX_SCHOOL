```python
number = 1
while number <= 3:
    print(number)
    number += 1
print("Loop finished")
```

- while문 안에서 반복을 중단할 때 사용하는 키워드 : break

```python
# while True :
#     print("무한루프")
#     break

# 일반적인 break 사용법
# while 옆은 True를 지정, while 내부에서 조건을 지정하여 break 문을 만나게 함.
num = 1
while True:
    print(num)
    if num >= 10:
        print(f"{num} 멈춤")
        break
    num += 1
    print("+1 추가하겠습니다")
    break
print(num)
```

```python
first_num = 1
second_num = 1

while True:
    first_num = int(input("첫 번째 정수 입력하세요 >>"))
    print(f"첫 번째 정수 입력하세요 >>{first_num}")
    second_num = int(input("두 번째 정수 입력하세요 >>"))
    print(f"두 번째 정수 입력하세요 >>{second_num}")
    
    
    sum = first_num + second_num
    
    if first_num == 0 and second_num == 0:
        print("모두 0을 입력하시면 안됩니다.")
        break
    else:
        print("두 정수의 합 : {}".format(sum))    
```

```python
# 두 정수를 입력받아 합을 구하는 프로그램
# 만약 두 정수 모두 0이면 프로그램을 종료합니다.

while True:
    # 첫 번째 정수 입력받기
    first_num = int(input("첫 번째 정수를 입력하세요 >> "))
    
    # 두 번째 정수 입력받기
    second_num = int(input("두 번째 정수를 입력하세요 >> "))
    
    # 만약 두 정수가 모두 0이면
    if first_num == 0 and second_num == 0:
        print("두 숫자 모두 0을 입력하셨습니다. 프로그램을 종료합니다.")
        break  # while문을 빠져나옴
    
    # 두 정수의 합을 계산
    sum = first_num + second_num
    
    # 결과 출력
    print("두 정수의 합은: {}".format(sum))
```

```python
# 현재 몸무게를 입력받아 정수로 변환
weight_now = int(input("현재 몸무게를 입력하세요 (kg) >> "))
print(f"현재 몸무게 : {weight_now}kg")  # 현재 몸무게 출력
# 목표 몸무게를 입력받아 정수로 변환
weight_goal = int(input("목표 몸무게를 입력하세요 (kg) >> "))
print(f"목표 몸무게 : {weight_goal}kg")  # 목표 몸무게 출력
# 주차(week) 변수 초기화
week = 1

# 현재 몸무게가 목표 몸무게보다 클 때 반복
while weight_now > weight_goal:
    # 이번 주 감량한 몸무게를 입력받음
    lose = int(input(f"{week}주차 감량 몸무게를 입력하세요 >>"))
    weight_now -= lose  # 현재 몸무게에서 감량한 몸무게를 뺌

    print(f"{week}주차 감량 몸무게 : {lose}kg")  # 이번 주 감량 결과 출력
    week += 1  # 주차 증가
print(f"{weight_goal}kg 도달! 축하합니다!")  # 목표 달성 메시지 출력



```
