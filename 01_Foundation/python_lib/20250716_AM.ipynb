```python
# 개발자의 남녀 비율 확인하기
# data.value_counts("성별")
data["성별"].value_counts()
```

```python
# 1. 개발자 전체 평균 월급 계산하기
# 2. 국적별 평균 월급 계산하기
# 3. 학력별 평균 월급 계산하기
# 4. 성별 기준 평균 월급 계산하기
data
```

```python
# 연봉데이터 -> 월급 -> 원화로 환산 (환율 1,386원=> ￦1386/＄1 )
data["월급"] = data["연봉"]/12 * 1386
# astype() 함수를 사용하여 월급을 정수형으로 변환
data["월급"] = data["월급"].astype(int)
data
```

```python
# 개발자 전체 평균 월급 계산하기
data["월급"].mean()
```

```python
# 국적별 평균 월급 계산하기 unique() 함수사용
data["국적"].unique()
```

```python
# 한국인의 월급 평균 계산하기 => 불리언 인덱싱 활용
# data[data["국적"] == "South Korea"]["월급"].mean()
korean = data[data["국적"] == "South Korea"]
korean["월급"].mean()
```

```python
# 학력에따른 월급 평균을 구해보자. 
edu = data.groupby(by= "학력").agg({"월급": "mean"})
edu

# 성별에 따른 월급 평균을 구해보자
gender = data.groupby(by= "성별").agg({"월급": "mean"})
gender

nation = data.groupby(by= "국적").agg({"월급": "mean"})
nation
```
