### 모듈
- 변수나 함수 또는 클래스를 모아 놓은 파일
- 다른 파이썬 프로그램에서 불러와 사용할 수 있게끔 만든 파이썬 파일
- 모듈은 다른 사람이 이미 만들어 놓은 모듈을 사용할 수도 있고 직접 만들어서 사용할 수도 있음
- 파이썬에서 사용할 수 있는 모듈은 확장자가 .py파일
- module : .py 파일 (ex 서류한장)
- package : module이 여러개 들어있는  폴더 (ex 보고서 하나)
- library : module + package  여러개 들어있는 도구의 모음 (ex 보고서가 모여있는 책장)




```python
# 모듈 import
import module
```

```python
# 모듈 호출법 : module.function name()
# 덧셈 함수호출
module.cal_add(4,7)
```

```python
# 뺄셈 함수호출
module.cal_sub(6,1)
```
