### Q3 아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오(10점)
- 데이터:
    - list1 = [1, 30, 7, 9, 17, 3, 5, 23, 44]
- 실행결과: 
    - 짝수 2개, 홀수 7개

```python
# 홀짝 구분 개수 세기
list1 = [1, 30, 7, 9, 17, 3, 5, 23, 44]

even_count = 0
odd_count = 0


for i in range(len(list1)):
    if list1[i]%2 == 0:
        even_count += 1
    else:
        odd_count += 1
        
print(f"짝수 {even_count}개, 홀수 {odd_count}개")
```

### Q7. 국내증시 시가총액 사이트의 결과 중, 코스피 상위 50개 기업의 종목명, 현재가, 시가총액 3가지를 수집하고, 데이터프레임으로 제작하는 프로그램을 작성하시오.

- 사이트 링크 : https://finance.naver.com/sise/sise_market_sum.naver
-  ※ 사용할 라이브러리 : requests, beautifulSoup, pandas




```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

# 클라이언트가 서버에 요청하고 요청받은 결과값 response 변수에 저장
response = req.get("https://finance.naver.com/sise/sise_market_sum.naver")
response

# lxml parser를 사용해서 text를 html형태로 변환
soup = bs(response.text, "lxml")
soup

# 종목명 선택자: #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(2) > a
# 현재가 선택자: #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(3)
# 시가총액 선택자: #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(7)

info_name = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(2) > a")
info_price = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(3)")
info_value = soup.select("#contentarea > div.box_type_l > table.type_2 > tbody > tr > td:nth-child(7)")

company_list = []
shareprice_list = []
companyvalue_list = []

for i in range(len(info_name)) : 
    company_list.append(info_name[i].text)
    shareprice_list.append(info_price[i].text)
    companyvalue_list.append(info_value[i].text)

kor_Corp_Info = pd.DataFrame({
    "종목명": company_list,
    "현재가": shareprice_list,
    "시가총액": companyvalue_list 
}) 

display(kor_Corp_Info)



```

### 08. 제시된 링크에서 스클롤링을 진행하고 모든 상품(오디오)의 이름, 가격, 평점을 수집하여 데이터프레임으로 결과값을 저장하시오.

```python
# package import
from selenium import webdriver as wd 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs

import time
import pandas as pd

driver = wd.Chrome()
driver.get("https://www.lge.co.kr/home-audio")


# 스클롤링 
body = driver.find_element(By.CSS_SELECTOR,"body")

while True:
    first = driver.page_source
    body.send_keys(Keys.END)
    time.sleep(0.5) 

    second = driver.page_source
    time.sleep(0.5)  

    if (first == second):
        print("더 이상 스크롤을 내릴 수 없습니다.")
        break  
    else:
        print("스크롤을 내렸습니다.")


product_name = driver.find_elements(By.CSS_SELECTOR,"#container > section.PlpPcContainer_container__L8y3U > section.PlpPcContainer_list_cont__tBF9C > section.CommonPcListUnitProduct_list_unit_product__AeuOT > ul > li > div.CommonPcListUnitProduct_title_wrap__4P_gA.undefined > a")
display(product_name)
product_price = driver.find_elements(By.CSS_SELECTOR, "#container > section.PlpPcContainer_container__L8y3U > section.PlpPcContainer_list_cont__tBF9C > section.CommonPcListUnitProduct_list_unit_product__AeuOT > ul > li > div.CommonPcListUnitProduct_price_wrap__52oXD > div > p > span.current_price > strong")
display(product_price)
product_reputation = driver.find_elements(By.CSS_SELECTOR, "#container > section.PlpPcContainer_container__L8y3U > section.PlpPcContainer_list_cont__tBF9C > section.CommonPcListUnitProduct_list_unit_product__AeuOT > ul > li > div.CommonPcListUnitProduct_title_wrap__4P_gA.undefined > div.product-card-title_product-card-title__CwsCk > div > span.product-card-title_rating > span.product-card-title_score")
display(product_reputation)

product_name_list = []
product_price_list =[]
product_reputation_list = []

for i in  range(len(product_name)):
     product_name_list.append(product_name[i].text),
     product_price_list.append(product_price[i].text),
     product_reputation_list.append(product_reputation[i].text)

LG_Product_Info = pd.DataFrame({
    "제품": product_name_list,
    "가격": product_price_list,
    "평점": product_reputation_list
})

display(LG_Product_Info)
```
