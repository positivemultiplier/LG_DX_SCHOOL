```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 한글 인코딩
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# 그 외 여러가지 그래프

# 넘파이로 랜덤 배열을 만들어보자
x = np.random.randint(6, 45, size=100)
y = np.random.randint(1, 51, size=100)

# 산점도 그래프
plt.scatter(x, y, color='skyblue',alpha=0.5 , marker='o') # -> 산점도 그래프
plt.xlim(0, 50)
plt.ylim(0, 80)
plt.title('산점도 그래프')
plt.show()
```

```python
# 파이차트
lab = ['a', 'b', 'c', 'd']
val = [30, 20, 40, 10]
col = ['#fcd0f5', '#e0d9fa', '#c8fdd7', '#e3faab']
plt.pie(val, labels=lab, autopct='%.1f%%', startangle=90, colors=col, counterclock=False)
plt.title('파이차트')
plt.axis('equal')
plt.show()
```

```python
# 카테고리화된 항목에대한 막대그래프 그리기
name = ['a', 'b', 'c', 'd']
value = [30, 20, 40, 10]
plt.bar(name, value, color='skyblue', alpha=0.5)
plt.title('막대그래프')
plt.show()
```

```python
# 수평 막대그래프
df = pd.DataFrame({
    'name': ['a', 'b', 'c', 'd'],
    'value': [30, 20, 40, 10]
})
plt.barh(df['name'], df['value'], color='skyblue', alpha=0.5)
plt.title('막대그래프 (DataFrame)')
plt.show()
```
