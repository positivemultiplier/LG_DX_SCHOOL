```python
driver = wd.Chrome()
driver.get("https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=lg")

```

```python
# ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜¤ê¸°ìœ„í•´ ë¬´í•œ ìŠ¤í¬ë¡¤ ì‹¤í–‰ !!! ë¡œë”©ì„ ë‹¤ í•œ ìƒíƒœì—ì„œ seleniumì€ ì§„í–‰í•´ì•¼í•œë‹¤. 
body = driver.find_element(By.CSS_SELECTOR,"body")
while True :
    first = driver.page_source
    body.send_keys(Keys.END)
    time.sleep(1)
    second = driver.page_source
    if(first == second) :
        break
```
