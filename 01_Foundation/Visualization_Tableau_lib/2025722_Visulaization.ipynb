# 시각화란 무엇인가?
- 데이터를 그래프, 차트, 지도 등 시각적 요소로 표현해 정볼르 더 쉽고 빠르게 이해할 수 있도록 돕는 기법
- 필요성
    - 데이터 양의 폭발적 증가
    - 복잡한 데이터의 이해도 향상
    - 빠른 의사결정 지원
- 활용예시
    - 코로나 확산 경로 시각화
    - 기업 KPI(key Perfomance Indicator) 대시보드
    - 선거 결과
    - 기후 변화
    - 트렌드 분석 등
- 좋은 시각화의 조건
    - 명확성(Clarity): 핵심 메시지가 한눈에 보임
    - 간결성(Simplicity): 불필요한 요소 제거 
    - 정확성(Accuracy): 데이터 왜곡 없이 표현
    - 일관성(Consistency): 색상, 축, 폰트 동일
    - 스토리텔링(Narrative): 데이터에 맥락 제공

## Table + au 태블로란?
- 전처리 기능도 있지만, numpy, pandas 등으로 전처리 후 시각화에 집중하는 것이 일반적
- 주요기능
    - 다양한 데이터 연결: Excel, SQL, Google Sheets 등
    - 막대, 선, 파이, 히트맵, 지도 등 다양한 차트
    - 대시보드 및 스토리 기능: 보고서 형태구성
    - 필터, 정렬, 조건부 서식 등 상호작용 기능
