```python
#    indexing  vs slicing
# Quize  : 10,20,30,40,50

# 1st   arr4 = np.array([10,20,30,40,50])
arr4 = np.array([10,20,30,40,50])

# 2nd numpy arange function(start, end, step)  
arr4 = np.arange(10, 51, 10)

arr4

```

```python
#     ! 
# indexing => [index number]
# 1st :  [0] (indexing) (printing) 
# step1 : indexing
a = arr4[0]
# step2 : printing
print(a)

# 2nd :    (indexing) (printing) 
# 10, 30    => index number 0,2
#       
# step1 : indexing
b = arr4[[0,2]]
# step2 : printing
print(b)

# 3rd :  5  (slicing) (printing)
# [row ][start index num(default : 0) : end index num(default : unlimit ) : step (default : +1)]
# step1 : slicing
c = arr4[0:5:1]
# step2 : printing
print(arr4[0:5:1])

# 4th :    (boolean indexing) (printing)
#  => [T/F ] => True  
# True  . 
# step1 : boolean indexting
d = arr4[arr4 > 30]
# step2 : printing
print(d)


# asdjkhasd
```
