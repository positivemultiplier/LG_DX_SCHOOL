````python
# BigData ìˆ˜ì—…ì—ì„œ ìì£¼ ì‚¬ìš©í•˜ëŠ” pandas ì˜ˆì‹œ ì½”ë“œ

# 1. ìƒ˜í”Œ ë°ì´í„° ìƒì„±
import pandas as pd

# DataFrame ìƒì„±
data = {
    'ì´ë¦„': ['ê¹€ì² ìˆ˜', 'ì´ì˜í¬', 'ë°•ë¯¼ìˆ˜', 'ìµœìˆ˜ì§„'],
    'ë‚˜ì´': [25, 30, 35, 28],
    'ì ìˆ˜': [85, 92, 78, 95],
    'ì§€ì—­': ['ì„œìš¸', 'ë¶€ì‚°', 'ì„œìš¸', 'ëŒ€êµ¬']
}
df = pd.DataFrame(data)
print("=== ì›ë³¸ ë°ì´í„° ===")
print(df)
print()

# 2. ê¸°ë³¸ ì •ë³´ í™•ì¸
print("=== ë°ì´í„° ì •ë³´ ===")
print(f"ë°ì´í„° í¬ê¸°: {df.shape}")
print(f"ì»¬ëŸ¼ íƒ€ì…: {df.dtypes.to_dict()}")
print()

# 3. ê¸°ìˆ í†µê³„ëŸ‰
print("=== ê¸°ìˆ í†µê³„ëŸ‰ ===")
print(df.describe())
print()

# 4. ë°ì´í„° ì„ íƒ ë° ì¡°íšŒ
print("=== ë°ì´í„° ì„ íƒ ì˜ˆì‹œ ===")
print("ì ìˆ˜ ì»¬ëŸ¼:", df['ì ìˆ˜'].tolist())
print("ì²« 2í–‰:", df.head(2))
print("ì ìˆ˜ 90 ì´ìƒ:", df[df['ì ìˆ˜'] >= 90])
print()

# 5. ì •ë ¬
print("=== ì •ë ¬ ì˜ˆì‹œ ===")
print("ì ìˆ˜ ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ:")
print(df.sort_values('ì ìˆ˜', ascending=False))
print()

# 6. ê·¸ë£¹ë³„ ì§‘ê³„
print("=== ê·¸ë£¹ë³„ ì§‘ê³„ ===")
print("ì§€ì—­ë³„ í‰ê·  ì ìˆ˜:")
print(df.groupby('ì§€ì—­')['ì ìˆ˜'].mean())
print()

# 7. ì§‘ê³„ í•¨ìˆ˜
print("=== ì§‘ê³„ í•¨ìˆ˜ ===")
print(f"í‰ê·  ì ìˆ˜: {df['ì ìˆ˜'].mean():.2f}")
print(f"ìµœê³  ì ìˆ˜: {df['ì ìˆ˜'].max()}")
print(f"ìµœì € ì ìˆ˜: {df['ì ìˆ˜'].min()}")
print(f"ì´ ì¸ì›: {df['ì ìˆ˜'].count()}")
````

````python
# ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë° ë°ì´í„° ë³€í™˜ ì˜ˆì‹œ

# ê²°ì¸¡ì¹˜ê°€ í¬í•¨ëœ ë°ì´í„° ìƒì„±
import numpy as np

missing_data = {
    'ì´ë¦„': ['í™ê¸¸ë™', 'ê¹€ì˜ìˆ˜', 'ë°•ì² ìˆ˜', 'ì´ë¯¼ì •'],
    'ì ìˆ˜1': [85, np.nan, 90, 78],
    'ì ìˆ˜2': [92, 88, np.nan, 85],
    'ë“±ê¸‰': ['A', 'B', 'A', 'C']
}
df_missing = pd.DataFrame(missing_data)
print("=== ê²°ì¸¡ì¹˜ í¬í•¨ ë°ì´í„° ===")
print(df_missing)
print()

# ê²°ì¸¡ì¹˜ í™•ì¸
print("=== ê²°ì¸¡ì¹˜ í™•ì¸ ===")
print("ê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(df_missing.isnull().sum())
print()

# ê²°ì¸¡ì¹˜ ì²˜ë¦¬
print("=== ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ===")
# 1) ê²°ì¸¡ì¹˜ ì œê±°
df_dropped = df_missing.dropna()
print("ê²°ì¸¡ì¹˜ ì œê±° í›„:")
print(df_dropped)
print()

# 2) ê²°ì¸¡ì¹˜ í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸° (ìˆ«ìí˜• ì»¬ëŸ¼ë§Œ í‰ê· ê°’ìœ¼ë¡œ ì±„ìš°ê¸°)
df_filled = df_missing.copy()
for col in ['ì ìˆ˜1', 'ì ìˆ˜2']:
    df_filled[col] = df_filled[col].fillna(df_filled[col].mean())
print("ê²°ì¸¡ì¹˜ í‰ê· ìœ¼ë¡œ ì±„ìš´ í›„:")
print(df_filled)
print()

# ë°ì´í„° ë³€í™˜ ì˜ˆì‹œ
print("=== ë°ì´í„° ë³€í™˜ ===")
# apply í•¨ìˆ˜ ì‚¬ìš©
df_filled['ì ìˆ˜_í‰ê· '] = df_filled[['ì ìˆ˜1', 'ì ìˆ˜2']].apply(lambda x: x.mean(), axis=1)
print("ì ìˆ˜ í‰ê·  ì¶”ê°€:")
print(df_filled)
print()

# map í•¨ìˆ˜ ì‚¬ìš© (ë“±ê¸‰ì„ ìˆ«ìë¡œ ë³€í™˜)
grade_map = {'A': 90, 'B': 80, 'C': 70}
df_filled['ë“±ê¸‰_ì ìˆ˜'] = df_filled['ë“±ê¸‰'].map(grade_map)
print("ë“±ê¸‰ì„ ì ìˆ˜ë¡œ ë³€í™˜:")
print(df_filled)
print()

# ì „ì¹˜ (í–‰ê³¼ ì—´ ë°”ê¾¸ê¸°)
print("=== ì „ì¹˜ (.T) ===")
print("ì „ì¹˜ ì „:")
print(df_filled[['ì ìˆ˜1', 'ì ìˆ˜2']].head(3))
print("\nì „ì¹˜ í›„:")
print(df_filled[['ì ìˆ˜1', 'ì ìˆ˜2']].head(3).T)
````

````python
# íŒŒì¼ ì…ì¶œë ¥ ì˜ˆì‹œ (BigData ìˆ˜ì—…ì—ì„œ ì¤‘ìš”!)

# 1. CSV íŒŒì¼ë¡œ ì €ì¥
print("=== CSV íŒŒì¼ ì €ì¥ ===")
# ì•ì„œ ë§Œë“  ë°ì´í„°ë¥¼ CSVë¡œ ì €ì¥
df_filled.to_csv('sample_data.csv', index=False, encoding='utf-8')
print("sample_data.csv íŒŒì¼ë¡œ ì €ì¥ ì™„ë£Œ")
print()

# 2. CSV íŒŒì¼ ì½ê¸°
print("=== CSV íŒŒì¼ ì½ê¸° ===")
# df_loaded = pd.read_csv('sample_data.csv', encoding='utf-8')
# print("ì €ì¥ëœ íŒŒì¼ ì½ê¸°:")
# print(df_loaded.head())
print("íŒŒì¼ ì½ê¸° ì˜ˆì‹œ: pd.read_csv('íŒŒì¼ëª….csv')")
print()

# 3. ìì£¼ ì‚¬ìš©í•˜ëŠ” íŒŒì¼ ì½ê¸° ì˜µì…˜ë“¤
print("=== ìì£¼ ì‚¬ìš©í•˜ëŠ” íŒŒì¼ ì½ê¸° ì˜µì…˜ ===")
print("""
# ê¸°ë³¸ ì‚¬ìš©ë²•
df = pd.read_csv('íŒŒì¼ëª….csv')

# ì¸ì½”ë”© ì§€ì • (í•œê¸€ íŒŒì¼)
df = pd.read_csv('íŒŒì¼ëª….csv', encoding='utf-8')

# íŠ¹ì • ì»¬ëŸ¼ë§Œ ì½ê¸°
df = pd.read_csv('íŒŒì¼ëª….csv', usecols=['ì»¬ëŸ¼1', 'ì»¬ëŸ¼2'])

# ì²« ë²ˆì§¸ í–‰ì„ í—¤ë”ë¡œ ì‚¬ìš©í•˜ì§€ ì•Šì„ ë•Œ
df = pd.read_csv('íŒŒì¼ëª….csv', header=None)

# êµ¬ë¶„ì ì§€ì • (íƒ­ìœ¼ë¡œ êµ¬ë¶„ëœ íŒŒì¼)
df = pd.read_csv('íŒŒì¼ëª….txt', sep='\t')

# íŠ¹ì • í–‰ ê°œìˆ˜ë§Œ ì½ê¸°
df = pd.read_csv('íŒŒì¼ëª….csv', nrows=1000)
""")
````

````python
# BigData ìˆ˜ì—… í•µì‹¬: ê·¸ë£¹ë³„ ì§‘ê³„ ë° ê³ ê¸‰ í™œìš©

# ë” ë³µì¡í•œ ìƒ˜í”Œ ë°ì´í„° ìƒì„±
big_data = {
    'ì§€ì—­': ['ì„œìš¸', 'ë¶€ì‚°', 'ì„œìš¸', 'ëŒ€êµ¬', 'ë¶€ì‚°', 'ì„œìš¸', 'ëŒ€êµ¬', 'ë¶€ì‚°'],
    'ì—°ë„': [2022, 2022, 2023, 2022, 2023, 2023, 2023, 2022],
    'ë§¤ì¶œ': [1000, 800, 1200, 600, 900, 1100, 700, 850],
    'ê³ ê°ìˆ˜': [50, 40, 60, 30, 45, 55, 35, 42],
    'ì¹´í…Œê³ ë¦¬': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B']
}
df_big = pd.DataFrame(big_data)
print("=== BigData ìƒ˜í”Œ ë°ì´í„° ===")
print(df_big)
print()

# 1. ë‹¨ì¼ ê·¸ë£¹ë³„ ì§‘ê³„
print("=== ë‹¨ì¼ ê·¸ë£¹ë³„ ì§‘ê³„ ===")
print("ì§€ì—­ë³„ í‰ê·  ë§¤ì¶œ:")
print(df_big.groupby('ì§€ì—­')['ë§¤ì¶œ'].mean())
print()

# 2. ë‹¤ì¤‘ ê·¸ë£¹ë³„ ì§‘ê³„
print("=== ë‹¤ì¤‘ ê·¸ë£¹ë³„ ì§‘ê³„ ===")
print("ì§€ì—­ë³„, ì—°ë„ë³„ ë§¤ì¶œ í•©ê³„:")
print(df_big.groupby(['ì§€ì—­', 'ì—°ë„'])['ë§¤ì¶œ'].sum())
print()

# 3. ë‹¤ì–‘í•œ ì§‘ê³„ í•¨ìˆ˜ ë™ì‹œ ì ìš©
print("=== ë‹¤ì–‘í•œ ì§‘ê³„ í•¨ìˆ˜ ë™ì‹œ ì ìš© ===")
result = df_big.groupby('ì§€ì—­')['ë§¤ì¶œ'].agg(['sum', 'mean', 'count', 'min', 'max'])
print(result)
print()

# 4. ì—¬ëŸ¬ ì»¬ëŸ¼ì— ëŒ€í•œ ì§‘ê³„
print("=== ì—¬ëŸ¬ ì»¬ëŸ¼ì— ëŒ€í•œ ì§‘ê³„ ===")
multi_agg = df_big.groupby('ì§€ì—­').agg({
    'ë§¤ì¶œ': ['sum', 'mean'],
    'ê³ ê°ìˆ˜': ['sum', 'mean']
})
print(multi_agg)
print()

# 5. ì¡°ê±´ë¶€ í•„í„°ë§ + ì§‘ê³„
print("=== ì¡°ê±´ë¶€ í•„í„°ë§ + ì§‘ê³„ ===")
print("ë§¤ì¶œ 800 ì´ìƒì¸ ë°ì´í„°ë§Œ ì§€ì—­ë³„ ì§‘ê³„:")
filtered_result = df_big[df_big['ë§¤ì¶œ'] >= 800].groupby('ì§€ì—­')['ë§¤ì¶œ'].mean()
print(filtered_result)
print()

# 6. í”¼ë²— í…Œì´ë¸” (BigData ë¶„ì„ì—ì„œ ìì£¼ ì‚¬ìš©)
print("=== í”¼ë²— í…Œì´ë¸” ===")
pivot_result = df_big.pivot_table(
    values='ë§¤ì¶œ', 
    index='ì§€ì—­', 
    columns='ì—°ë„', 
    aggfunc='sum', 
    fill_value=0
)
print(pivot_result)
print()

# 7. ë°ì´í„° ìš”ì•½ ë° ì¸ì‚¬ì´íŠ¸
print("=== ë°ì´í„° ìš”ì•½ ë° ì¸ì‚¬ì´íŠ¸ ===")
print(f"ì´ ë§¤ì¶œ: {df_big['ë§¤ì¶œ'].sum():,}ì›")
print(f"í‰ê·  ë§¤ì¶œ: {df_big['ë§¤ì¶œ'].mean():.2f}ì›")
print(f"ìµœê³  ë§¤ì¶œ ì§€ì—­: {df_big.loc[df_big['ë§¤ì¶œ'].idxmax(), 'ì§€ì—­']}")
print(f"ê°€ì¥ ë§ì€ ë°ì´í„°ê°€ ìˆëŠ” ì§€ì—­: {df_big['ì§€ì—­'].value_counts().index[0]}")
print()

# 8. ì—°ë„ë³„ ì„±ì¥ë¥  ê³„ì‚° (ê³ ê¸‰ í™œìš©)
print("=== ì—°ë„ë³„ ì„±ì¥ë¥  ê³„ì‚° ===")
yearly_sales = df_big.groupby('ì—°ë„')['ë§¤ì¶œ'].sum()
print("ì—°ë„ë³„ ì´ ë§¤ì¶œ:")
print(yearly_sales)
if len(yearly_sales) > 1:
    growth_rate = ((yearly_sales[2023] - yearly_sales[2022]) / yearly_sales[2022]) * 100
    print(f"2022ë…„ ëŒ€ë¹„ 2023ë…„ ì„±ì¥ë¥ : {growth_rate:.2f}%")
````
