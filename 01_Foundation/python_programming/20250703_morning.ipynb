# 딕셔너리(dictionary)란?
- dictionary는 단어 그대로 해석하면 사전이라는 뜻
- "people"이라는 단어는 "사람", "baseball"이라는 단어는 "야구"에 부합되듯이 dictionary는 key와 value를 한쌍으로 갖는 자료형
- 딕셔너리 타입은 immutable한 key와 mutable한 value로 맵핑되어 있는  ★순서가 없는(index 존재X key로 접근)  집합
- immutable : str, int, float, tuple
- mutable : dict, list

````python
# 딕셔너리 정의
# {key: value, key: value, ...}
# 한쌍의 요소를 구분해주는 기호 : 콤마(,)
a = {'name': '박성화'}
# 타입
type(a)
````

````python
b = {1 : '박성화', 2: '이진헌', 3: '서희창', 999: '외계인'}
# b
print(b[999])
````

## 1. 컬럼별 리스트를 값으로 하는 딕셔너리  dic{} in list[] VS list[] in dic 
(**예시:** pandas DataFrame과 비슷한 구조)

```python
member = {
    'name': [...],
    'phone': [...],
    'email': [...]
}
```

### **특징**
- 각 컬럼(속성)별로 데이터를 한 번에 다룰 수 있음
- 인덱스를 기준으로 한 행(row)에 접근 가능
- 데이터프레임(pandas)로 변환이 매우 쉬움

### **활용 예시**
- **데이터 분석**:  
  여러 명의 전화번호만, 이메일만 한 번에 추출할 때  
  ```python
  print(member['phone'])  # 전화번호 전체 리스트
  ```
- **pandas DataFrame 변환**:  
  ```python
  import pandas as pd
  df = pd.DataFrame(member)
  print(df)
  ```
- **일괄 처리**:  
  예를 들어, 모든 이메일에 도메인 붙이기 등 컬럼 단위 연산이 편리

### **단점**
- 개별 멤버의 모든 정보를 한 번에 보기 위해서는 인덱스를 맞춰야 함  
  (예: 2번째 멤버의 name, phone, email을 모두 출력하려면 각 리스트의 1번 인덱스를 사용)

## 2. 멤버별 딕셔너리의 리스트  
(**예시:** 데이터베이스의 레코드 구조와 유사)

```python
members = [
    {'name': ..., 'phone': ..., 'email': ...},
    ...
]
```

### **특징**
- 각 멤버의 모든 정보를 하나의 딕셔너리로 묶어서 관리
- 특정 멤버의 모든 속성에 쉽게 접근 가능
- 조건 검색, 필터링, 멤버 추가/삭제가 용이

### **활용 예시**
- **특정 멤버 검색**:  
  ```python
  for m in members:
      if m['name'] == '이진헌':
          print(m['phone'], m['email'])
  ```
- **조건부 필터링**:  
  예를 들어, 경영학 전공만 추출  
  ```python
  mgmt = [m for m in members if m['major'] == '경영학']
  ```
- **데이터베이스 저장/불러오기**:  
  JSON, MongoDB 등과 구조가 유사해 직관적으로 저장/불러오기 가능

### **단점**
- 모든 멤버의 특정 속성만 한 번에 추출하려면 리스트 컴프리헨션 등 추가 작업 필요  
  ```python
  emails = [m['email'] for m in members]
  ```

## **요약 표**

| 스타일                   | 장점                                 | 단점                                   | 대표 활용 |
|--------------------------|--------------------------------------|----------------------------------------|-----------|
| 컬럼별 리스트 딕셔너리   | 컬럼 단위 연산, DataFrame 변환 쉬움  | 멤버 단위 접근 불편                    | 데이터분석, 일괄처리, pandas |
| 멤버별 딕셔너리 리스트   | 멤버 단위 접근, 조건 검색 쉬움       | 컬럼 단위 연산 불편                    | 검색, 필터링, DB저장, JSON |

## **실무 팁**
- **데이터 분석/통계/머신러닝** → 컬럼별 리스트 구조 + pandas
- **웹/앱 개발, DB 저장, 조건 검색** → 멤버별 딕셔너리 리스트 구조
