## 조건문
- 상황에 따라 실행 흐름을 다르게 하는 문법

```python
a = int(input("숫자를 입력하세요 >> "))
b = int(input("숫자를 입력하세요 >> "))

if a>=b :
    print(f'{a}"이(가)" {b}"보다 더 큽니다"')
else :
    print(f'{b}"이(가)" {a}"보다 더 큽니다"')
```

```python
try:
    num1 = int(input("첫 번째 정수를 입력하세요 >> "))
except :
    print("정확한 수를 입력하세요>> ")
    
num2 = int(input("두 번째 정수를 입력하세요 >> "))

if num1>num2 :
    print(f'{num1} 이(가) {num2} 보다 {num1-num2} 만큼 더 큽니다.')
    print('{} 이(가) {} 보다 {} 만큼 더 큽니다.'.format(num1,num2,num1-num2))
elif num1<num2 :
    print(f'{num2} 이(가) {num1} 보다 {num2-num1} 만큼 더 큽니다.')
elif num1 == num2 :
    print(f'{num1} 과 {num2} 이 똑같습니다')
    print('{} 과 {} 이 똑같습니다.'.format(num1,num2)) 
else :
    print('정확한 값을 입력하세요')
```

```python
print('---[SM COFFEE]---')

menu = {
    1: {'name': '에스프레소', 'price': 600},
    2: {'name': '아메리카노', 'price': 1000},
    3: {'name': '카페라떼', 'price': 1300}
}

try:
    money_amount = int(input('금액을 입력하세요 >> '))

    # 메뉴 안내 메시지 동적 생성
    menu_message = ', '.join(
        [f"[{num}] {info['name']}({info['price']}원)" for num, info in menu.items()]
    )
    print('커피를 고르세요')
    select_menu = int(input(f"{menu_message} >> "))

    if select_menu in menu:
        price = menu[select_menu]['price']
        if money_amount < price:
            print('잔액이 부족합니다')
        else:
            change = money_amount - price
            print(f"거스름돈은 {change}원 입니다.")
    else:
        print('메뉴 선택 시 정확한 번호를 입력하세요.')

except ValueError:
    print('금액 입력 시 숫자를 입력하세요.')

```

```python
worktime = int(input('근무시간 입력 >> '))

if worktime <= 8:
    worktime_base = worktime
    worktime_exceed = 0
else : 
    worktime_base = 8
    worktime_exceed = worktime - 8


wage = [ 
    ['기본시급', 10000],
    ['초과시급', 10000 * 1.5]
]

wage_total = (worktime_base * wage[0][1]) + (worktime_exceed * wage[1][1])

print(f'총  지급액은 {int(wage_total)}원 입니다.')
```

```python
subject = ['DB', '전자계산기 구조', '운영체제', '데이터통신', '소프트웨어공학']
scores = []
score0 = int(input(f"{subject[0]} >> "))
score1 = int(input(f"{subject[1]} >> "))
score2 = int(input(f"{subject[2]} >> "))
score3 = int(input(f"{subject[3]} >> "))
score4 = int(input(f"{subject[4]} >> "))

scores = [score0, score1, score2, score3, score4]
score_sum = scores[0]+scores[1]+scores[2]+scores[3]+scores[4]

if score0 < 8 or score1 < 8 or score2 < 8 or score3 < 8 or score4 < 8:
    print("불합격입니다!")
elif sum(scores) < 60:
    print(f"{score_sum}점 불합격입니다!")
else:
    print(f"{score_sum}점 합격입니다!")
```
