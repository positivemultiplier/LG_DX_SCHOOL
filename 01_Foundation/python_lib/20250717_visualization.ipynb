# Visualization 시각화
- 많은 양의 자료를 한눈에 볼 수 있도록 도표나, 차트 등으로 정리하는 것
- 시각화를 통해 데이터의 특징을 쉽게 파악 가능
- matplotlib 라이브러리를 활용한다.
- MATLAB + plot + library

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글 인코딩 하는법 한번해두면 계속 적용됨 
# runtime configuration = 런타임 설정
plt.rc("font", family="Malgun Gothic")
```

```python
# line chart 그리기
# 데이터 형식 => list 형태로 넣기 => plot은 선을 의미한다. 
# 데이터를 하나만 넣어주면 axis = 0 인가봐. y축으로 간다, x축은 incex처럼 자동으로 부여된다. 
# 경제학처럼 독립변수와 종속변수가 바뀌었네. 
print("y축에 데이터 넣을께요, x축은 자동으로 들어갑니다 ") 
y = [2, 4, 6, 8]
plt.plot(y)
plt.show()

# x축 좌표의 폰트사이즈를 바꾸고싶을때 
print("x축 폰트사이즈 키워주세요 !!!")
plt.xticks(fontsize=20)
plt.show()
```

```python
# x축과 y축에 데이터 넣기
print("x축과 y축에 모두 데이터 넣을께요")
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)
plt.show()
```

```python
# 뻘짓 3차원은 안된다. 
# x축과 y축에 데이터 넣기
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [1, 3, 5, 7]
plt.plot(x, y, z)
plt.show()
```

```python
# 넘파이를 이용한 그래프 그리기
x = [1, 2, 3, 4]
y = np.array([2, 4, 6, 8]) # 차원만 맞으면 그려짐

# 속성에 대한 특징 !! 
# ls를  이용한 점선처리 ls = line style => "-- "
# lw를  이용한 선굵기 조절 lw = line width => 5
# color를 이용한 색상조절 color = "red"
# marker = "o" => 점을 찍어주는 역할
# mfc 를 이용한 marker의 색상 조절 mfc = marker face color => "blue" 
plt.plot(x, y, ls=":", lw=5, color="red", marker="o", mfc="blue")
plt.show()
```

# 사고유형 차대차 사건 중 사상자를 확인해보기 => 발생지 시도 기준으로 

```python
data.head()
```

```python
# 사고유형 대분류 인덱싱
data["사고유형_대분류"]

# 사고유형 대분류별 사고 건수 세기
data["사고유형_대분류"].value_counts()

# 사고유형 대분류에서 차대차인것만 가져오기
data[data["사고유형_대분류"] == "차대차"]
# 사고유형 대분류에서 차대차인것만 가져오기 + 숫자세기 
data[data["사고유형_대분류"] == "차대차"].value_counts()

# 사고유형 대분류에서 차대차인것만 가져오기  => 변수에저장하기
car = data[data["사고유형_대분류"] == "차대차"]

# 필요한 정보만 인덱싱하기 > 사상자수,발생지시도 
data1 = car[["사상자수", "발생지시도"]]
data1
```

```python
# 발생지시도별로 사고 건수 세기
result = data1.groupby("발생지시도").sum()
result
```

```python
# 선 그래프 그려보기 (X => 지역, y => 사상자수)

x = result.index
y = result["사상자수"].values



plt.plot(x, y, ls=":")
plt.title("대물 사고 발생건수 by 지역", loc = "left", pad = 10)
plt.xlabel("지역", loc = "right", labelpad = 10)
plt.ylabel("사상자수" )
plt.show()
```

```python
# bar 그래프를 그려보자
plt.bar(x, y)


# bar 그래프 꾸미기.
plt.title("차대차 교통사고의 사상자 수 ", pad = 100, loc = "center" )

#x축 지역
plt.xlabel("지역", loc= "center", labelpad = 100)

#y축 사상자 => 글 바꾸기.
plt.ylabel("사상자",loc = "center" , labelpad = 100, rotation =0)

# 그래프 보여줘 
plt.show()
```

# 파이차트(pie chart) 그리기!
## 파이차트 속성 
- 

```python
# 파이차트  라벨 그리기 = > 리스트로 만들어야한다.
labels = ["A", "B", "C", "D"]

# 파이차트 비율 구분 => 점유울은 정수로. 
sizes = [15, 30, 45, 10]

explode = [0.1, 0, 0, 0]

# 파이차트 그리기 속성 => 1st: 점유비율 2nd : 라벨이름
plt.pie(sizes, labels= labels,  colors = ["red", "blue", "green", "yellow"], autopct="%.1f%%", startangle=90,counterclock=False, explode=[0.5, 0, 0, 0]  )

## 파이차트 속성 percent 표시하기 .  autopct="%.1f%%" => 소숫점 몇재짜리까지 표시할께. 
## colors = ["red", "blue", "green", "yellow"] => 색상 지정
## startangle=90 => 시작각도 90도로 설정
## counterclock=False => 시계방향으로 그리기
## explode=[0.1, 0, 0, 0] => 첫번째 조각을 살짝 떼어내기

plt.title("파이차트")

plt.show()
```

# 카테고리화된 항목에대한 막대그래프 그리기

```python
# 카테고리화된 항목에 대한 막대 그래프 그리기
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]
color=["red", "blue", "green", "yellow"]

plt.bar(categories, values, color=["red", "blue", "green", "yellow"], width=0.4)

# 막대 그래프 제목 설정
plt.title("카테고리화된 수직 막대그래프")


plt.show()

# 순서를 바꾸고 싶다면
plt.bar(categories[::-1], values[::-1], color=["red", "blue", "green", "yellow"], width=0.4)

plt.title("카테고리화된 수직 막대그래프 (순서 변경)")
plt.show()



# 막대그래프 수평으로 그리기
plt.barh(categories, values, color=["red", "blue", "green", "yellow"], height=0.4)

# 수평 막대 그래프 제목 설정
plt.title("카테고리화된 수평 막대그래프")

plt.show()


# 순서를 바꾸고 싶다면
plt.barh(categories[::-1], values[::-1], color=color[::-1], height=0.4)
plt.title("카테고리화된 수평 막대그래프 (순서 변경)")
plt.show()

```
