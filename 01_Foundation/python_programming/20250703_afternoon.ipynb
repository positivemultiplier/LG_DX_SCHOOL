# HTML(HyperText Markup Language)


- Element = tag + content

- Attribute(속성)  Value(값)


- auto rename tag
- live server

## 1) HTML (HyperText Markup Language)
	- Markup : 표시하겠다.
	- 역할 : 페이지의 뼈대를 담당(영역을 배치 -> 글자,문단,제목..)
	- 핵심 : 태그로 소통한다.
	- <b> </b> : 태그 -> 모습을 결정한다.
	- 내용 : 컨텐츠 -> 사용자에게 보여줄 내용
	- 태그를 꾸며주는 내용 : 속성 -> align, href
	- 태그 + 컨텐츠 + 속성 : element(요소)
 
## 2) h태그 : 제목을 작성할 때 사용하는 태그
	* 특징 : 1번~6번까지 존재한다
	* 사용법 : 크기가 작아질수록 소제목을 의미한다.
 
## 3) p태그 : 문단을 작성할 때 사용하는 태그
	* 특징 : 위아래 공백, 한줄 전체를 차지(자동줄바꿈)
	* 사용법 : 많은 양의 텍스를 작성할 때 문단제작
 
## 4) span태그 : 특정 문자,문장을 표기할 때 사용하는 태그
	* 특징 : 아무런 기능이 없는 공태그
	* 사용법 : 특정 내용을 공간으로 감싸줄때 활용한다.
 
## 5) b,strong태그 : 글자의 크기는 같지만 두께를 두껍게 만든다.
	* 특징 : b태그는 기능이 없는 디자인, strong태그는 악세트를 가졌다.
	* 사용법 : 최근에는 대부분 strong태그를 활용한다.
 
## 6) a태그 : 특정 내용과 사이트,파일을 연결할 때 활용하는 태그
	* 특징 : 특정 내용과 연결시키기 위해서 반드시 목적지 선정이 필요하다.
	* 사용법 : href속성을 활용한다.
 
## 7) 내일 배울 내용
	1) 위키만들기 실습
	2) 목록만들기
	3) 이미지태그 -> 경로
	4) 테이블태그
